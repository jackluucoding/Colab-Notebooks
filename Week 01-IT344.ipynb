```
import statistics 
import numpy as np
import pandas as pd
dataset =[9,14,15,20,25]

#Find Z-Score normalization 
from scipy import stats
z_score = stats.zscore(dataset)
print("The Z-score is: ",z_score)


#Find mean, mode, median, and standard deviation
mean = np.mean(dataset)
mode = stats.mode(dataset)
median = np.median(dataset)
standard = np.std(dataset)
print('Mean:', mean,'\nMode:', mode,'\nMedian:', median,'\nStandard Deviation:',standard )

#Find skewness
skewness = stats.skew(dataset)
print('skewness is:', skewness)


#Find Q1, Q2, Q3 
result = np.percentile(dataset, [25, 50, 75])
print('Q1, Q2, and Q3 (Respectively): ', result)
```
