```
import networkx as nx
import matplotlib.pyplot as plt

#create a graph with all chracters
GoT = nx.Graph()

#adding 20 characters
GoT.add_nodes_from(['Daenerys','Viserys','Howland','AerysII','Jaime','Rhaegar','Lyanna','Elia','Robert','JonArryn','RhaenysNAegon','Ned','Gregor','Rhaella','Rickard','Arthur','JonSnow','Benjen','Rhaegar','Brandon','Tywin'])


#--------------
#Abduction links 
Abduct = nx.DiGraph()                  
Abduct.add_edges_from([('Rhaegar', 'Lyanna')], label='abducted')


#---------------
#Served links
Serve = nx.DiGraph()
Serve.add_edges_from([('Arthur','AerysII'), ('Jaime','AerysII'), ('Gregor','Tywin')], label = 'served')


#-------------------
#Guardian links 
Guardian = nx.DiGraph()
Guardian.add_edges_from([('Ned', 'JonSnow'), ('JonArryn', 'Robert'), ('JonArryn','Ned')], label='guardian')



#--------------------
#Kill links 
Kill = nx.DiGraph()
Kill.add_edges_from([('Ned', 'Arthur' ), ('Gregor', 'RhaenysNAegon' ), ('Gregor','Elia'), ('Howland','Arthur'), ('AerysII','Brandon'), ('AerysII', 'Rickard'), ('Jaime','AerysII'), ('Robert','Rhaegar')], label='killed')

#---------------------
#Parent links
Parent = nx.DiGraph()
Parent.add_edges_from([('Tywin','Jaime'),('AerysII','Viserys'),('AerysII','Daenerys'),('Rhaella','Daenerys'),('Rhaella','Rhaegar'),('Rickard','Brandon'),('Rhaegar','RhaenysNAegon'),('Rhaegar','JonSnow'),('Lyanna','JonSnow'),('Rickard','Ned'),('Rickard','Benjen'),('Elia','RhaenysNAegon')], label='parent')



#---------------------
#Married/Engaged links
Married = nx.Graph()
Married.add_edges_from([('Lyanna','Robert'),('AerysII','Rhaella'),('Rhaegar','Elia')], label='married/engaged')
                       
#--------------------
#Allies links                    
Allies = nx.Graph()
Allies.add_edges_from([('Howland','Ned'), ('Tywin','Robert'),('Ned','Robert'),('JonArryn','Robert')], label='Allies')      


#------------------------
#Siblings links
Siblings = nx.Graph()
Siblings.add_edges_from([('Viserys','Daenerys'),('Viserys','Rhaegar'), ('AerysII','Rhaella'),('Daenerys','Rhaegar'), ('Brandon','Lyanna'),('Brandon','Ned'),('Brandon','Benjen'),('Benjen','Lyanna'),('Benjen','Ned'),('Lyanna','Ned')],label='siblings')



plt.figure(1, figsize=(12,12))
pos = nx.circular_layout(GoT)
nx.draw(GoT, pos, node_color='grey', node_size=5, with_labels=True)


nx.draw_networkx_edges(Abduct, pos, edge_color='deeppink')
nx.draw_networkx_edges(Serve, pos, edge_color='purple')
nx.draw_networkx_edges(Kill, pos, edge_color='black')
nx.draw_networkx_edges(Guardian, pos, edge_color='black')
nx.draw_networkx_edges(Parent, pos, edge_color='blue')
nx.draw_networkx_edges(Married, pos, edge_color='pink')
nx.draw_networkx_edges(Allies, pos, edge_color='orange')
nx.draw_networkx_edges(Siblings, pos, edge_color='green')

#Extrating labels from the list of edges 
Abduct_labels = nx.get_edge_attributes(Abduct,'label')
Serve_labels = nx.get_edge_attributes(Serve,'label')
Kill_labels = nx.get_edge_attributes(Kill,'label')
Guardian_labels = nx.get_edge_attributes(Guardian,'label')
Parent_labels = nx.get_edge_attributes(Parent,'label')
Married_labels = nx.get_edge_attributes(Married,'label')
Allies_labels = nx.get_edge_attributes(Allies,'label')
Siblings_labels = nx.get_edge_attributes(Siblings,'label')


nx.draw_networkx_edge_labels(Abduct, pos, edge_labels = Abduct_labels, label_pos=0.2, font_color='deeppink')
nx.draw_networkx_edge_labels(Serve, pos, edge_labels=Serve_labels, label_pos=0.2,font_color='purple')
nx.draw_networkx_edge_labels(Kill, pos, edge_labels = Kill_labels, label_pos=0.2, font_color='red')
nx.draw_networkx_edge_labels(Guardian, pos, edge_labels = Guardian_labels, label_pos=0.2, font_color='pink')
nx.draw_networkx_edge_labels(Parent, pos, edge_labels=Parent_labels, label_pos=0.2,font_color='blue')
nx.draw_networkx_edge_labels(Married, pos, edge_labels=Married_labels, label_pos=0.2,font_color='purple')
nx.draw_networkx_edge_labels(Allies, pos, edge_labels=Allies_labels, label_pos=0.2,font_color='orange')
nx.draw_networkx_edge_labels(Siblings, pos, edge_labels=Siblings_labels, label_pos=0.2,font_color='brown')

plt.show()

for n in GoT.nodes():
      print('Number of actions that',n,'involved in:',Abduct.degree(n),Serve.degree(n),Guardian.degree(n),Kill.out_degree(n),Parent.degree(n), Married.degree(n), Allies.degree(n), Siblings.degree(n))           
```
