```
import pandas as pd
import numpy as np

from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler 
from time import time
from time import time

from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout

np.random.seed(1457)
```

```
from google.colab import files
uploaded = files.upload()
uploaded1 = files.upload()

#Browse to the folder where"cleanTitanicTest.csv" and "cleanTitanicTrain.csv" are located
```

```
import io
titanicTest = pd.read_csv(io.BytesIO(uploaded['cleanTitanicTest.csv']))
len(titanicTest)
```

```
import io
titanicTrain = pd.read_csv(io.BytesIO(uploaded1['cleanTitanicTrain.csv']))
len(titanicTrain)
```

```
predictorNames = ["Pclass", "Sex", "Age", "Fare"]

myScaler = StandardScaler() #scaling and centering 

trainPredictors=myScaler.fit_transform(titanicTrain[predictorNames].values)

trainTarget=titanicTrain['Survived']

trainTargetonehot = pd.get_dummies(titanicTrain['Survived']).values

testPredictors = myScaler.transform(titanicTest[predictorNames].values)
testPIDs = titanicTest['PassengerId']

start = time()

myModel = Sequential()
myModel.add(Dense (10, input_dim=4))
myModel.add(Dropout(0.1))
myModel.add(Dense(10))
myModel.add(Dense(output_dim=2))

myModel.add(Activation('softmax'))

myModel.compile(loss='binary_crossentropy', optimizer ='adagrad', metrics=['binary_accuracy'])

myModel.fit(trainPredictors, trainTargetonehot)

myModelPredictions = myModel.predict_classes(trainPredictors)

#print(classification_report(trainTarget, myModelPredictions))

predictedTarget = myModel.predict_classes(testPredictors)

mySubmission = pd.DataFrame({'PassengerID':
                            testPIDs, 'Survived' : predictedTarget})

mySubmission.to_csv("mySubmission.csv", index=False)
```

```


start = time()

myNewModel = Sequential()
myNewModel.add(Dense (3, input_dim=4))
myNewModel.add(Dropout(0.1))
myNewModel.add(Dense(10))
myNewModel.add(Dense(output_dim=2))


myNewModel.add(Activation('sigmoid'))

myNewModel.compile(loss='binary_crossentropy', optimizer ='adagrad', metrics=['binary_accuracy'])

myNewModel.fit(trainPredictors, trainTargetonehot)
#myNewModel.fit(trainPredictors, trainTargetonehot,  epochs=100, batch_size=32, validation_split=0.2, verbose=0)


myModelPredictions = myNewModel.predict_classes(trainPredictors)

print(classification_report(trainTarget, myModelPredictions))

predictedTarget = myNewModel.predict_classes(testPredictors)

myNewSubmission = pd.DataFrame({'PassengerID':
                            testPIDs, 'Survived' : predictedTarget})

myNewSubmission.to_csv("myNewSubmission.csv", index=False)
```
